### Momentum Trade Strategy based on Sp500

```python
import yfinance as yf 
import numpy as np 
import pandas as pd 
import datetime as dt
from pandas.tseries.offsets import MonthEnd

import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
tickers = df.Symbol.to_list()
```

```python
len(tickers)
```

```python
df2 = yf.download(tickers, start='2009-12-31',end='2020-12-31')
```

```python
df2
```

```python
#filter only for the adjusted close prices 
prices = df2['Adj Close']
```

```python
prices
```

```python
#transform index from string to datetime index 
prices.index = pd.to_datetime(prices.index)
```

```python
#claculate monthly returns 
mtl_ret = prices.pct_change().resample('M').agg(lambda x: (x+1).prod() - 1 if x.isnull().sum() == 0 else np.nan)
```

```python
mtl_ret
```

```python
#calculate past 11 month returns
past_11 = (mtl_ret+1).rolling(11).apply(np.prod)-1 
```

```python
past_11
```

```python
formation = dt.datetime(2010,12,31)
formation 
```

```python
end_measurement = formation - MonthEnd(1)
end_measurement
```

```python
ret_12 = past_11.loc[end_measurement]
ret_12
```

```python
ret_12 = ret_12.reset_index()
ret_12
```

```python
#set new decile column based on from percetage change in ten quantiles 
ret_12['decile'] = pd.qcut(ret_12.iloc[:,1],10,labels=False,duplicates='drop')
```

```python
ret_12
```

```python
winners = ret_12[ret_12.decile == 9]
losers = ret_12[ret_12.decile == 0]
```

```python
winners
```

```python
winnerret = mtl_ret.loc[formation + MonthEnd(1), mtl_ret.columns.isin(winners['index'])]
loserret = mtl_ret.loc[formation + MonthEnd(1), mtl_ret.columns.isin(losers['index'])]
```

```python
momentum_profit = winnerret.mean() - loserret.mean()
momentum_profit
```

```python
def momentum(formation):
    end_measurement = formation - MonthEnd(1)
    ret_12 = past_11.loc[end_measurement]
    ret_12 = ret_12.reset_index()
    ret_12['decile'] = pd.qcut(ret_12.iloc[:,1],10,labels=False,duplicates='drop')
    winners = ret_12[ret_12.decile == 9]
    losers = ret_12[ret_12.decile == 0]
    winnerret = mtl_ret.loc[formation + MonthEnd(1), mtl_ret.columns.isin(winners['index'])]
    loserret = mtl_ret.loc[formation + MonthEnd(1), mtl_ret.columns.isin(losers['index'])]
    momentum_profit = winnerret.mean() - loserret.mean()
    return momentum_profit
```

```python
momentum(formation)
```

```python
for i in range(12*10):
    print(formation + MonthEnd(i))
```

```python
profits = []
dates = []

for i in range(12*10):
    profits.append(momentum(formation + MonthEnd(i)))
    dates.append(formation+MonthEnd(i))
```

```python
data_set = {"dates":dates,"profits":profits}
df_neo = pd.DataFrame(data=data_set)
```
