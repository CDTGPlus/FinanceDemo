### Bollinger Bands

```python
import yfinance as yf 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = yf.download('MSFT', start='2021-01-01')
```

```python
df
```

```python
#calculate simple moving average 
df['SMA'] = df.Close.rolling(window=20).mean()
df
```

```python
df['stddev'] = df.Close.rolling(window=20).std()
```

```python
df
```

```python
df['Upper'] = df.SMA + 2* df.stddev
df['Lower'] = df.SMA - 2* df.stddev

df 
```

```python
#Define buying signal; when lower band is above the close price
df['Buy_Signal'] = np.where(df.Lower > df.Close, True, False) 
```

```python
#Define selling signal, when the upper band is below the close price 
df['Sell_Signal'] = np.where(df.Upper < df.Close, True, False) 
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
plt.rcParams["figure.figsize"] = (12,6)
plt.plot(df[['Close','SMA','Upper','Lower']])
plt.scatter(df.index[df.Buy_Signal],df[df.Buy_Signal].Close, marker='^', color='r')
plt.scatter(df.index[df.Sell_Signal],df[df.Sell_Signal].Close, marker='^', color='r')
plt.fill_between(df.index,df.Upper,df.Lower,color='grey',alpha=0.3)
plt.legend(['Close','SMA','Upper','Lower'])
plt.show()
```

```python
#Create a strategy through which a sell or a purchase is committed uppon the first respective signal along the time series 
buys = []
sells = []
open_pos = False

for i in range(len(df)):
    if df.Lower[i] > df.Close[i]:
        if open_pos == False:
            buys.append(i)
            open_pos = True 
    elif df.Upper[i] < df.Close[i]:
        if open_pos:
            sells.append(i)
            open_pos = False
```

```python
buys
```

```python
sells
```

```python
plt.rcParams["figure.figsize"] = (12,6)
plt.plot(df[['Close','SMA','Upper','Lower']])
plt.scatter(df.iloc[buys].index, df.iloc[buys].Close, marker='^', color='r')
plt.scatter(df.iloc[sells].index, df.iloc[sells].Close, marker='^', color='r')
plt.fill_between(df.index,df.Upper,df.Lower,color='grey',alpha=0.3)
plt.legend(['Close','SMA','Upper','Lower'])
plt.show()
```

```python
merged = pd.concat([df.iloc[buys].Close, df.iloc[sells].Close], axis=1)
merged
```

```python
merged.columns = ['Buys','Sells']
merged
```

```python
totalprofit = merged.shift(-1).Sells - merged.Buys
totalprofit
```

```python
#relative profits
relprofits = (merged.shift(-1).Sells - merged.Buys)/merged.Buys
relprofits
```

```python
relprofits.mean()
```
