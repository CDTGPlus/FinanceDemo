### Short Term Reversal

#### Buy prior month worst performing stocks and hold them for one month

```python
import yfinance as yf
import pandas as pd 
import datetime as dt 
import numpy as np 
from pandas.tseries.offsets import MonthEnd
```

```python
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/S%26P_100')[2]

tickers
```

```python
tickers = tickers.Symbol.to_list()
```

```python
tickers = ['BRK-B' if item == 'BRK.B' else item for item in tickers]
```

```python
df = yf.download(tickers,start='2009-01-01')
```

```python
df
```

```python
prices = df['Adj Close']
```

```python
prices
```

```python
prices.index = pd.to_datetime(prices.index)
```

```python
#relative price changes accumulated over a month
mtl_ret = prices.pct_change().resample('M').agg(lambda x: (x+1).prod() - 1 )
```

```python
mtl_ret
```

```python
formation = dt.datetime(2009,2,28)
formation
```

```python
ret_1 = mtl_ret.loc[formation - MonthEnd(1)]
ret_1
```

```python
ret_1.name = 'mtl_ret'
```

```python
ret_1 = pd.DataFrame(ret_1)
ret_1
```

```python
ret_1['decile'] = pd.qcut(ret_1.mtl_ret,10,labels=False)
ret_1
```

```python
losers = ret_1[ret_1.decile == 0].index
losers
```

```python
losret = mtl_ret.loc[formation,mtl_ret.columns.isin(losers)].mean()
losret
```

```python
def reversal(formation):
    ret_1 = mtl_ret.loc[formation - MonthEnd(1)]
    ret_1.name = 'mtl_ret'
    ret_1 = pd.DataFrame(ret_1)
    ret_1['decile'] = pd.qcut(ret_1.mtl_ret,10,labels=False,duplicates='drop')
    losers = ret_1[ret_1.decile == 0].index
    losret = mtl_ret.loc[formation,mtl_ret.columns.isin(losers)].mean()
    return losret
```

```python
formation_2 = dt.datetime(2009,3,31)
reversal(formation_2)
```

```python
mtl_ret
```

```python
returns = []
dates = []
for i in mtl_ret.index[2:]:
    returns.append(reversal(i))
    dates.append(i)
```

```python
returns
```

```python
frame = pd.DataFrame({'dates':dates,'returns':returns})
```

```python
frame
```

```python
frame.returns.mean()
```
